```python
#Split multi-country entries
countries = df['country'].str.split(', ')
countries.head()
```

```python
countries.explode().value_counts().head()
```

After splitting and exploding the country column, the United States remains the top producer of Netflix content. However, the ranking of other countries changes, highlighting the importance of properly handling multi-country entries to avoid misleading conclusions

```python
countries_exploded = df[['type', 'country']].dropna()
countries_exploded.head()
```

```python
countries_exploded['country'] = countries_exploded['country'].str.split(', ')
countries_exploded['country'].head()
```

```python
countries_exploded = countries_exploded.explode('country')
countries_exploded.head()
```

```python
country_type_count = countries_exploded.groupby(['country', 'type']).size().reset_index(name='count').sort_values('count', ascending=False)
country_type_count.head(10)
```

```python
type_by_year = df.groupby(['release_year', 'type']).size().reset_index(name='count').sort_values('release_year', ascending=False)
type_by_year.head(10)
```

```python
type_by_year['type'].value_counts()
```

Movies are released across more distinct years than TV Shows, indicating a longer or more consistent production history.

```python
country_type_count['type'].value_counts()
```

Movies are distributed across more countries than TV Shows, suggesting wider global availability.
